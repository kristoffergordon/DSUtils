```python
import numpy as np
import pandas as pd
from dsutils import plots
```

```python
rng = np.random.default_rng()
df = pd.DataFrame(rng.random((10, 5)), columns=['A', 'B', 'C', 'D', 'E'])
```

```python
plots.heatmap(df)
```

```python
plots.correlation_heatmap(df)
```

```python
# Create a Series with some random data
np.random.seed(0)
s = pd.Series(np.random.randn(1000))

# Display a histogram and a normal probability plot
plots.probplts(s)
```

```python
# Create two numpy arrays with some random data
np.random.seed(0)
y_test = np.random.rand(100)
y_pred = y_test + np.random.normal(0, 0.2, 100)

# Display the prediction error plot
plots.pred_error_plt(y_pred, y_test, rng_min=0, rng_max=100)
```

```python
# Create a DataFrame with some random data
np.random.seed(0)
df = pd.DataFrame({
    'A': np.random.choice(['Very Low', 'Low', 'Medium', 'High', 'Very High', ' Extreme'], 250),
    'B': np.random.choice(['Yes', 'No', 'Maybe'], 250),
    'C': np.random.rand(250)
})

# Display boxplots
plots.serial_boxplot(df, 'C')
```
